```python
import re

# Open the ACH file and read all lines
with open('WrongNacha.ach', 'r') as f:
    lines = f.readlines()

# Initialize total debit and credit amounts to zero
total_debit = 0.0
total_credit = 0.0

# Loop through each line in the file
for i, line in enumerate(lines):
    # Search for the Total Debit and Total Credit amounts
    match_debit = re.search(r'^8\d{16}.*\d{10}(\d{12})$', line)
    match_credit = re.search(r'^6\d{16}.*\d{10}(\d{12})$', line)

    # If a Total Debit amount is found, add it to the total debit amount
    if match_debit:
        total_debit += float(match_debit.group(1)) / 100.0

    # If a Total Credit amount is found, add it to the total credit amount
    if match_credit:
        total_credit += float(match_credit.group(1)) / 100.0

    # Search for the Batch Control information
    match_batch = re.search(r'^8\d{16}.*\d{6}(\d{8})\d{10}(\d{12})$', line)

    # If a Batch Control line is found
    if match_batch:
        # Get the total debit and credit amounts for the batch
        batch_debit = float(match_batch.group(1)) / 100.0
        batch_credit = float(match_batch.group(2)) / 100.0

        # Calculate the difference between the batch debit and credit amounts
        diff = batch_debit - batch_credit

        # If the difference is not zero, adjust the Total Debit or Total Credit amount accordingly
        if diff != 0:
            if diff > 0:
                total_credit += diff
            else:
                total_debit -= diff

            # Update the Total Debit or Total Credit amount in the Batch Control line
            lines[i] = re.sub(r'^8\d{16}.*(\d{10})\d{12}$', f'8{"0"*10}{int((total_debit+total_credit)*100):012}', lines[i])

# Print the Total Debit and Total Credit amounts
print(f'Total Debit: {total_debit:.2f}')
print(f'Total Credit: {total_credit:.2f}')

# Ask user for value to search and replace
search_value = input('Enter the value to search: ')
replace_value = input('Enter the value to replace with: ')

# Loop through each line in the file
for i, line in enumerate(lines):
    # Replace the search value with the replace value in the line
    lines[i] = line.replace(search_value, replace_value)

# Write the updated lines back to the ACH file
with open('WrongNacha.ach', 'w') as f:
    f.writelines(lines)
```
