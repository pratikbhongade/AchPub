```python
# Define index for Batch Control Total Receiving DFI Identification field in Batch Control Record
batch_total_rdi_index = 10
file_total_rdi_index = 21

# Define additional indexes for debit and credit amounts
batch_total_debit_index = 20
batch_total_credit_index = 32
file_total_debit_index = 31
file_total_credit_index = 43

# Ask the user for the value to search
search_value = input("Enter the value to search: ")

# Read the contents of the ACH file
with open('Nacha.ach', 'r') as f:
    lines = f.readlines()

# Initialize batch control and file control total amounts
batch_total_rdi = {}
batch_total_debit = {}
batch_total_credit = {}
file_total_rdi = 0
file_total_debit = 0
file_total_credit = 0
first_file_control = True

# Create a new list to store the updated lines
updated_lines = []

# Initialize variables for batch and sequence number to remove
batch_to_remove = ""
sequence_to_remove = ""
remove_batch_control = False

# Loop through the lines of the ACH file and search for the record
for line in lines:
    record_type = line[0]

    if record_type == '5':
        # Extract the batch number from the Batch Header Record
        current_batch_number = line[87:94].strip()
        # Initialize the batch total receiving DFI identification sum and debit/credit amounts for the current batch
        batch_total_rdi[current_batch_number] = 0
        batch_total_debit[current_batch_number] = 0
        batch_total_credit[current_batch_number] = 0

        # Check if the current batch matches the batch to remove and reset the remove variables
        if current_batch_number == batch_to_remove:
            batch_to_remove = ""
            sequence_to_remove = ""

    elif record_type == '6':
        # Check if the search value is present in the Receiving DFI Identification field
        rdi = line[rdi_index:rdi_index + 8]

        if search_value in rdi:
            # Print the line where the search value is found
            print("Found search value in line: {}".format(line.strip()))

            # Print the batch information
            entry_detail_sequence_number = line[87:94].strip()
            print("Search value is present in Batch {} and Sequence No {}.".format(current_batch_number, entry_detail_sequence_number))

            # Ask the user if they want to remove this record
            remove_record = input("Do you want to remove this record? (y/n): ")
            if remove_record.lower() == 'y':
                # Set the batch and sequence number to remove
                batch_to_remove = current_batch_number
                sequence_to_remove = entry_detail_sequence_number

                # Check if the batch contains only one entry record
                if len(batch_total_rdi) == 1:
                    remove_batch_control = True

                continue  # Skip adding the line to the updated_lines list

    if current_batch_number == batch_to_remove and record_type == '8':
        # Check if the current line is the batch control record for the batch to be removed
        # If so, skip adding it to the updated_lines list
        continue

    elif current_batch_number == batch_to_remove and record_type == '7':
        # Check if the current line is the batch header record for the batch to be removed
        # If so, skip adding it to the updated_lines list
        continue

    elif record_type == '6' and current_batch_number == batch_to_remove:
        # Check if the current line is the record to be removed
        # By skipping it, we effectively remove it from the updated_lines list
        if line[87:94].strip() == sequence_to_remove:
            # Subtract the debit and credit amounts from the batch and file control totals
            batch_total_debit[current_batch_number] -= int(line[11:23])
            batch_total_credit[current_batch_number] -= int(line[21:33])
            file_total_debit -= int(line[11:23])
            file_total_credit -= int(line[21:33])
            continue

    # Add the line to the updated_lines list
    updated_lines.append(line)

    if record_type == '8':
        # Update the batch control total receiving DFI identification sum, debit and credit amounts for the current batch
        if current_batch_number in batch_total_rdi:
            batch_total_rdi[current_batch_number] += int(line[batch_total_rdi_index:batch_total_rdi_index + 10])
        if current_batch_number in batch_total_debit:
            batch_total_debit[current_batch_number] += int(line[batch_total_debit_index:batch_total_debit_index + 12])
        if current_batch_number in batch_total_credit:
            batch_total_credit[current_batch_number] += int(line[batch_total_credit_index:batch_total_credit_index + 12])

    elif record_type == '9' and first_file_control:
        # Update the file control total receiving DFI identification sum, debit and credit amounts
        file_total_rdi += int(line[file_total_rdi_index:file_total_rdi_index + 10])
        file_total_debit = int(line[file_total_debit_index:file_total_debit_index + 12])
        file_total_credit = int(line[file_total_credit_index:file_total_credit_index + 12])
        first_file_control = False

# Remove the batch control line if necessary
if remove_batch_control:
    for i, line in enumerate(updated_lines):
        record_type = line[0]
        if record_type == '8' and line[87:94].strip() == batch_to_remove:
            del updated_lines[i]
            break  # Stop removing after the first matching batch control line

# Recalculate the file control total receiving DFI identification sum

# Update the file control total receiving DFI identification sum, debit, and credit amounts
file_total_rdi = sum(int(val) for val in batch_total_rdi.values())
file_total_debit = sum(int(val) for val in batch_total_debit.values())
file_total_credit = sum(int(val) for val in batch_total_credit.values())

# Convert the file control total receiving DFI identification sum, debit, and credit amounts to strings
file_total_rdi_str = str(file_total_rdi).zfill(10)
file_total_debit_str = str(file_total_debit).zfill(12)
file_total_credit_str = str(file_total_credit).zfill(12)

# Find the index positions of the file control record fields
file_total_debit_index = 31
file_total_credit_index = 43
file_total_rdi_index = 21

for i, line in enumerate(updated_lines):
    record_type = line[0]
    if record_type == '9':
        # Update the file control record with the updated total debit amount
        updated_line = line[:file_total_debit_index] + file_total_debit_str + line[file_total_debit_index + 12:]

        # Update the file control record with the updated total credit amount
        updated_line = updated_line[:file_total_credit_index] + file_total_credit_str + updated_line[file_total_credit_index + 12:]

        # Update the file control record with the updated total receiving DFI identification sum
        updated_line = updated_line[:file_total_rdi_index] + file_total_rdi_str + updated_line[file_total_rdi_index + 10:]

        updated_lines[i] = updated_line
        break  # Stop updating after the first file control record

# Write the updated lines to the ACH file
with open('CorrectNacha.ach', 'w') as f:
    f.writelines(updated_lines)

# Print the updated file control total receiving DFI identification sum, debit, and credit amounts
print("Updated File Control Total Receiving DFI Identification Sum:", file_total_rdi)
print("Updated File Control Total Debit Entry Dollar Amount:", file_total_debit)
print("Updated File Control Total Credit Entry Dollar Amount:", file_total_credit)

# Print the batch control total receiving DFI identification sum, debit and credit amounts for each batch
for batch_number, total_rdi in batch_total_rdi.items():
    total_debit = batch_total_debit.get(batch_number, 0)
    total_credit = batch_total_credit.get(batch_number, 0)
    print("Batch Control Total Receiving DFI Identification Sum for Batch {}: {}".format(batch_number, total_rdi))
    print("Batch Control Total Debit Amount for Batch {}: {}".format(batch_number, total_debit))
    print("Batch Control Total Credit Amount for Batch {}: {}".format(batch_number, total_credit))

# Print the file control total receiving DFI identification sum, debit and credit amounts
print("File Control Total Receiving DFI Identification Sum: {}".format(file_total_rdi))
print("File Control Total Debit Amount: {}".format(file_total_debit))
print("File Control Total Credit Amount: {}".format(file_total_credit))
```
